```python
# importing the required libraries
import pandas as pd
import re
```

```python
# reading CSV
df = pd.read_csv('sample_addresses.csv', header = None)
# Printing the dataframe
df.drop(columns=[1], inplace =True)
```

```python
# changing data to lowercase
df = df.apply(lambda x: x.astype(str).str.lower())
df = df.drop([5,15,18])
# df
```

```python
full_address_list = []
street_number_name_list = []
post_code_list = []
state_list = []
suburb_list = []

for index, row in df.iterrows():
  full_address= row[0]
  print(full_address)

  # Fetching the street address from the full address
  # full_address = "11 - 19 belair close park ridge south qld 4125"
  street_number_name = re.compile("\d{1,4} [\w\s\-\/]{1,20}(?:street|st|avenue|ave|road|rd|way|highway|hwy|square|sq|trail|trl|drive|dr|court|ct|parkway|pkwy|circle|cir|boulevard|blvd|close)\W?(?=\s|$)")
  street_number_name_string = street_number_name.search(full_address).group(0)

  # Extracting remaining address
  all_except_street_number_name_string = street_number_name.split(full_address)[1]

  # Extracting Post Code
  post_code = re.compile(r"\D+")
  post_code = post_code.split(all_except_street_number_name_string)[1]

  #Extracting the state and suburb string
  suburb_state = all_except_street_number_name_string.split(post_code)[0]

  #Extracting State acronym
  state = re.compile("(vic|qld|act|wa|sa|nsw|tas)")
  state = state.search(suburb_state).group(0)

  #Extracting Suburb
  suburb = suburb_state.split(state)[0]

  full_address_list.append(full_address)
  street_number_name_list.append(street_number_name_string)
  post_code_list.append(post_code)
  state_list.append(state)
  suburb_list.append(suburb)
```

```python
df1 = pd.DataFrame({"Street":street_number_name_list,
                    "Suburb":suburb_list,
                    "State":state_list, 
                    "Postcode":post_code_list
                    })  
#showing data using dataframe              

df1 
```

```python
# saving the data to csv file after split
df1.to_csv('addresses.csv', index= False)
```
